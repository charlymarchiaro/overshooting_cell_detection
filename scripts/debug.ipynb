```python
from io import BytesIO
from zipfile import ZipFile
import pandas as pd
from IPython.display import display, Markdown, Latex
```

```python
src_data_folder = '..\\src_data\\traffica_insights_post_processing_jobs'
```

```python
src_file = 'traffica_insights_post_processing_job_89.zip'

def extract_zip(input_zip):
    input_zip=ZipFile(input_zip)
    return {name: input_zip.read(name) for name in input_zip.namelist()}

path = src_data_folder + '\\' + src_file

data = extract_zip(path)
```

```python
stats_summary_df = None
stats_total_df = None
stats_filtered_df = None
data_df = None
for key in data.keys():
    if key.startswith('stats'):
        stats_summary_df = pd.read_excel(BytesIO(data[key]), sheet_name='Resumen')
        stats_total_df = pd.read_excel(BytesIO(data[key]), sheet_name='Detalle (Total)')
        stats_filtered_df = pd.read_excel(BytesIO(data[key]), sheet_name='Detalle (Filtrado)')
    if key.startswith('data'):
        data_df = pd.read_csv(BytesIO(data[key]))
```

```python
display(Markdown('# ' + src_file))
display(Markdown('### Resumen'))
display(stats_summary_df)
display(Markdown('### Detalle (Total)'))
display(stats_total_df)
display(Markdown('### Detalle (Filtrado)'))
display(stats_filtered_df)
display(Markdown('### Datos'))
display(data_df)
```
